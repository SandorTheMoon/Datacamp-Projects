## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |

| `director` | Director of the show |

| `release_year` | Year of Netflix release |

| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
# Check for Null data
netflix_df.isnull().sum()
```

```python
# Filter only 'Movie' type
is_movie = netflix_df["type"] == "Movie"
movie_type = netflix_df[is_movie]
```

```python
# Filter movies from 1990s decade
is_1990s = (2000 > movie_type["release_year"]) & (movie_type["release_year"] >= 1990)
is_from_1990s = movie_type[is_1990s]
```

```python
# Get the list of frequent values for duration column
frequent_durations = is_from_1990s["duration"].mode()

# Get the top most frequent value from the rows
duration = frequent_durations.iloc[0]

print(f"The most frequent duration of movies during 1990s was {duration} minutes.")
```

```python
# Filter 'Action' movies
is_action = netflix_df["genre"] == "Action"
action_movies = netflix_df[is_action]
```

```python
# Filter Action movies from 1990s
is_1990s_action = (2000 > action_movies["release_year"]) & (action_movies["release_year"] >= 1990)
action_movies_from_1990s = action_movies[is_1990s_action]
```

```python
# Filter short action movies from 1990s (less than 90 minutes)
is_short = action_movies_from_1990s["duration"] < 90
short_action_movies = action_movies_from_1990s[is_short]
```

```python
# Get the count of short action movies from 1990s
# Based on the count of rows represented by shape[0]
short_movie_count = short_action_movies.shape[0]

print(f"The count of short actiom movies from 1990s is {short_movie_count}.")
```
